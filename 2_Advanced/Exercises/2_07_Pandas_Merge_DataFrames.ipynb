```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df['job_posted_month'] = df['job_posted_date'].dt.to_period('M')
df_us_jobs = df[df['job_country'] == 'United States']
df_monthly_postings = df_us_jobs.groupby('job_posted_month').size().reset_index(name='postings_count')
merged_df = pd.merge(df_us_jobs, df_monthly_postings, on='job_posted_month', how='right')
merged_df.head()
```

**Explanation:**

2. Use the dt accessor to extract the year from the job_posted_date column and create a new column job_posted_month by extracting the month from the job_posted_date column.
3. Create a DataFrame df_us_jobs that filters job postings in the United States.
4. Create a DataFrame df_monthly_postings that counts the number of job postings for each month using the groupby() method.
5. Use the merge() function to combine these two DataFrames on the 'job_posted_month' column using a right join.
6. Display the first 5 rows of the resulting DataFrame using the head() method.

## ðŸŸ¨ Merge Non-Remote Jobs and Salary (2.7.2)

```python
df_not_remote_jobs = df[(df["job_work_from_home"] == False)].copy()
df_not_remote_salary = df_not_remote_jobs.groupby("job_title_short")["salary_year_avg"].mean().reset_index(name = "salary_not_remote_avg")
merge_df = pd.merge(df_not_remote_jobs, df_not_remote_salary, on = "job_title_short", how = "right")
merge_df.head()
```

```python
df_not_remote_jobs = df[df['job_work_from_home'] == False]
df_not_remote_salary = df_not_remote_jobs.groupby('job_title_short')['salary_year_avg'].mean().reset_index()
df_not_remote_salary.rename(columns={'salary_year_avg': 'salary_not_remote_avg'}, inplace=True)
merged_df = pd.merge(df_not_remote_jobs, df_not_remote_salary, on='job_title_short', how='right')

merged_df.head()
```

## ðŸŸ¨ Merge Company Salary and Remote Count (2.7.3)

```python
df_company_salary = df.groupby("company_name")["salary_year_avg"].mean()
df_company_remote = df[(df["job_work_from_home"] == True)].copy()
df_company_remote.groupby("company_name").size().reset_index(name = "remote_count")
df_merge = pd.merge(df_company_salary, df_company_remote, on="company_name", how="inner")
df_merge.head()
```

```python
df_company_salary = df.groupby('company_name')['salary_year_avg'].mean().reset_index()
df_company_remote = df[df['job_work_from_home'] == True].groupby('company_name').size().reset_index(name='remote_count')
merged_df = pd.merge(df_company_salary, df_company_remote, on='company_name', how='inner')
merged_df.head()
```
