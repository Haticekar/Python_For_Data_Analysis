```python
import matplotlib.pyplot as plt

countries = ['USA', 'Germany', 'Canada', 'UK', 'India']
job_counts = [1000, 800, 600, 500, 900]
median_salaries = [120000, 110000, 100000, 95000, 30000]

# Scatter plot çiz
plt.scatter(job_counts, median_salaries)

# Her noktaya ülke adını yaz
for i in range(len(countries)):
    plt.text(job_counts[i], median_salaries[i], countries[i], fontsize=9)

# Başlık ve etiketler
plt.title("Job Count vs Median Salary by Country")
plt.xlabel("Number of Job Postings")
plt.ylabel("Median Salary (USD)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sahte örnek veri seti
data = {
    'job_title': ['Data Scientist', 'Backend Dev', 'Frontend Dev', 'ML Engineer', 'Data Analyst'],
    'job_count': [120, 90, 100, 70, 110],
    'median_salary': [130000, 115000, 105000, 125000, 95000],
    'avg_experience': [3.5, 4.0, 2.5, 4.5, 2.0],
    'job_category': ['Data', 'Software', 'Software', 'Data', 'Data']
}

df = pd.DataFrame(data)

# Kategorileri sayıya çevir (Data: 0, Software: 1)
color_map = {'Data': 'blue', 'Software': 'orange'}
df['color'] = df['job_category'].map(color_map)


plt.figure(figsize=(10, 6))

# Noktaları çiz: 
# x=job_count, y=median_salary, 
# s=avg_experience*100 (boyut), c=color (renk)
plt.scatter(
    x=df['job_count'],
    y=df['median_salary'],
    s=df['avg_experience'] * 100,  # tecrübeye göre nokta boyutu
    c=df['color'],
    alpha=0.7,
    edgecolors='black'
)

# Etiketleri ekle
for i in range(len(df)):
    plt.text(df['job_count'][i], df['median_salary'][i], df['job_title'][i],
             fontsize=9, ha='center', va='bottom')

# Eksen ve başlık
plt.xlabel("Number of Job Postings")
plt.ylabel("Median Yearly Salary (USD)")
plt.title("Job Postings vs Median Salary (with Experience & Category)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 🔍 Scatter Plot Nedir?

Bir scatter plot (dağılım grafiği), iki değişkenin birbiriyle nasıl ilişkili olduğunu görselleştirmek için kullanılır.

1. X ekseni bir değişkeni (örneğin iş ilanı sayısı)
2. Y ekseni başka bir değişkeni (örneğin yıllık ortalama maaş) gösterir.
3. Her nokta bir gözlemi temsil eder (örneğin bir ülke, bir iş, bir kişi vs.)

### 🧠 Scatter Plot – Püf Noktaları

| Parametre           | Açıklama                                                                 |
|---------------------|--------------------------------------------------------------------------|
| `plt.scatter(x, y)` | Temel scatter plot (dağılım grafiği) çizer                               |
| `s`                 | Nokta büyüklüğü (örneğin tecrübeye göre ölçeklenebilir)                  |
| `c`                 | Nokta rengi (tek renk ya da liste olabilir)                              |
| `alpha`             | Saydamlık (0 ile 1 arasında değer alır, 1 = tam opak)                    |
| `marker`            | Noktanın şekli (örneğin `'o'`, `'^'`, `'x'`, `'*'`)                      |
| `edgecolors`        | Noktanın kenar çizgisi rengi (örneğin `'black'`)                         |
| `plt.text()`        | Belirli bir koordinata yazı (etiket) eklemek için kullanılır             |
| `plt.xlabel()`      | X eksenine başlık ekler                                                  |
| `plt.ylabel()`      | Y eksenine başlık ekler                                                  |
| `plt.title()`       | Grafiğe başlık ekler                                                     |
| `plt.tight_layout()`| Grafik ögelerinin üst üste binmesini önleyip otomatik düzenleme yapar    |
